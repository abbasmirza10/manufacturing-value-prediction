```python
import numpy as np #Relevant imports
import pandas as pd
```

```python
csv_url = "https://data.nber.org//nberces/nberces5818v1/nberces5818v1_n2012.csv"
df = pd.read_csv(csv_url) #Reading in the dataset #print(df.isnull().sum()) - checking for missing values in the dataset
df_cleaned = df.dropna() # Dropping data elements with missing values to clean the dataset
df_cleaned
```

### SVM
Support Vector Machine

**Explanation of model**:

A SVM, or support vector machine, is a supervised ML algorithm that allows you to classify data through finding an optimal line or hyperplane. This optimal division maximizes the distance between each class and maximizes the margin between the closest data points of opposite classes.
In our case, we are splitting the data into training and testing data and are employing a SVM model with kernel "Linear".



```python
# SVM
import pandas as pd
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
X = df_cleaned.drop(columns=["vadd"])
Y = df_cleaned["vadd"]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
svm_model = svm.SVR(kernel='linear')
svm_model.fit(X_train, y_train)
svm_predictions = svm_model.predict(X_test)
mse = mean_squared_error(y_test, svm_predictions)
r2 = r2_score(y_test, svm_predictions)
print(svm_predictions)
print(mse)
print(r2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6)) # Scatter plot to compare actual and predicted values
plt.scatter(y_test, svm_predictions, alpha=0.5)
plt.title('Comparison of Actual and Predicted Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)  # Line showing perfect predictions
plt.show()
```

### OLS Regression
Ordinary Least Squares Regression

**Explanation of model**:

OLS Regression, or ordinary least squares regression, is a ML model used to predict the unknown parameters in a linear regression model. OLS minimizes the sum of the squared differences between the observed dependent variable values and values predicted by the model. In our research report, we have used OLS to predict the VADD, or value added, by the predictors in the dataset.


```python
# OLS Regression
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
X_OLS = df_cleaned.drop(columns=["vadd"])
Y_OLS = df_cleaned["vadd"]
X_train, X_test, y_train, y_test = train_test_split(X_OLS, Y_OLS, test_size=0.2, random_state=42)
ols_model = LinearRegression()
ols_model.fit(X_train, y_train)
ols_predictions = ols_model.predict(X_test)
```

In the code above, we perform OLS regression on the dataset. We are trying to predict VADD (Total value added in $1m) as seen by the X_OLS and the Y_OLS variables. After declaring the X_OLS, and the Y_OLS variables, the data is split into training and testing data, which is x_train, x_test, y_train, and y_test respectively. The model is then trained and fit on the data and results are evaluated by predicting the VADD using X_test as the independent variable. The result of this ML model is a line that fits the linear regression model to the training data.

ols_predictions is a line that generates the predictions for the target variable VADD using the trained linear regression model as well as the predictors.

```python
plt.scatter(y_test, ols_predictions, color='teal') # predicted vs actual vadd
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Actual VADD')
plt.ylabel('Predicted VADD')
plt.title('Actual vs Predicted VADD Values')
plt.show()
```

This graph plots the ols_predictions and is a visual representation of the actual VADD values obtained from the dataset versus the predicted VADD values resulting from the OLS regression model. The black line represents the values where the actual VADD is equal to the predicted VADD. The green colored dots shows the predicted VADD values in relation to this black line.

Overall, this plot visually assess how well the linear regression model performs in predicting the target variable 'VADD'. Because the green colored dots are all scattered around the black line, the OLS regression model is an accurate predictor of the "VADD" target variable.

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, ols_predictions)
r2 = r2_score(y_test, ols_predictions)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
residuals = y_test - ols_predictions # Plot residuals
plt.scatter(y_test, residuals, color='purple')
plt.xlabel('Actual VADD')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.axhline(y=0, color='black', linestyle='--')
plt.show()
```

### LASSO
Least Absolute Shrinkage and Selection Operator

**Explanation of model**:

Lasso, or Least Absolute Shrinkage and Selection Operator, is a machine learning model that is used primarily for regression and feature selection. It's goal is to reduce the sum of the squared differences between the actual values and the predicted values and also penalizing the absolute value of the coefficients. In this ML model, we shrink the coefficients of the less important features towards zero, which allows us to end with a much simpler model with fewers predictors. This effectively helps us with feature selection when choosing what predictors are most siginificant with regard to the data.




```python
#Lasso
from sklearn.linear_model import Lasso # Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
X_Lasso = df_cleaned.drop(columns=["vadd"])
Y_Lasso = df_cleaned["vadd"]
X_train, X_test, y_train, y_test = train_test_split(X_Lasso, Y_Lasso, test_size=0.2, random_state=42)
lasso_model = Lasso(alpha=0.1, max_iter = 10000) # Create a Lasso regression model
lasso_model.fit(X_train, y_train) # Fit the Lasso model to the training data
y_pred_lasso = lasso_model.predict(X_test) # Make predictions on the testing data
mse_lasso = mean_squared_error(y_test, y_pred_lasso) # Evaluate the Lasso model using Mean Squared Error
print("Mean Squared Error (Lasso):", mse_lasso)
coefficients_df = pd.DataFrame({'Feature': X_train.columns, 'Coefficient': lasso_model.coef_})
print("Coefficients (Lasso):\n", coefficients_df) # Print coefficients table
```

```python
n = len(X_Lasso) # Define the number of observations and predictors
k = len(X_Lasso.columns)
R_squared = lasso_model.score(X_Lasso, Y_Lasso) # Calculate the R-squared value
adjusted_R_squared = 1 - ((1 - R_squared) * (n - 1) / (n - k - 1))
print("Adjusted R-squared:", adjusted_R_squared)
y_pred = lasso_model.predict(X_test)
residuals = y_test - y_pred
RSS = np.sum(residuals**2)
print("Residual Sum of Squares (RSS):", RSS)
```

The actual vs. predicted plot compares the actual VADD values with the corresponding predicted values generated by the regression model. Each line represents a data point, with the blue line representing the actual VADD values and the red line representing the predicted VADD values. By visually comparing these lines, one can assess the model's accuracy in capturing the underlying patterns and variability in the data. A close alignment between the actual and predicted lines indicates a good fit, while deviations suggest areas where the model may need improvement, such as underfitting or overfitting.

```python
# Residual Plot: This plot shows the residuals against the predicted values. It helps to check if the residuals have a pattern or if they are randomly distributed.
residuals = y_test - y_pred_lasso
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_pred_lasso, y=residuals, alpha=0.8)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residual Plot')
plt.xlabel('Predicted VADD')
plt.ylabel('Residuals')
plt.show()
```

The residual plot depicts the distribution of residuals, which are the differences between the actual and predicted VADD values. Ideally, residuals should be randomly distributed around the horizontal line at y=0, indicating that the model's predictions are unbiased. ```python
#Feature Importance Plot: For Lasso regression, you can visualize the coefficients of the predictors to see which predictors are most important for predicting the target variable.
plt.figure(figsize=(10, 6))
sns.barplot(x=X_Lasso.columns, y=lasso_model.coef_)
plt.title('Feature Importance (Lasso)')
plt.xlabel('Features')
plt.ylabel('Coefficient')
plt.xticks(rotation=45)
plt.show()
```

The feature importance plot displays the coefficients of the predictor variables (EMP, PAY, VSHIP, INVEST) in the Lasso regression model. Larger coefficients suggest stronger associations with the target variable, while coefficients close to zero indicate less influence. ```python
#Coefficients Path Plot: For Lasso regression, plot the path of coefficients as the regularization strength (alpha) varies. This can help in understanding the effect of regularization on feature selection.
import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning)
alphas = np.logspace(-4, 4, 100)
lasso_coefs = []
for alpha in alphas:
    lasso_model = Lasso(alpha=alpha)
    lasso_model.fit(X_train, y_train)
    lasso_coefs.append(lasso_model.coef_)
plt.figure(figsize=(10, 6))
plt.plot(alphas, lasso_coefs)
plt.xscale('log')
plt.title('Lasso Coefficients Path')
plt.xlabel('Alpha (Regularization Strength)')
plt.ylabel('Coefficient Values')
plt.show()
```

### 10 Fold CV
10 Fold Cross Validation

**Explanation of model**:

10-fold CV, or 10 fold cross validation, is a ML model where you first need to split the data into 10 parts. The model is then trained 10 times, where one part is used a testing set and the remaining nine parts are used a validation set (which are different each time). For each iteration, the model is evaluated on the test set using performance metrics such as accuracy, precision, recall, and the final model evaluation is an average of these statistics. For our data, we are using the y variable as the VADD and the remaining variables as our predictor.




```python
#10 fold cv
from sklearn.model_selection import KFold, learning_curve
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
X = df_cleaned.drop(columns=["vadd"])  # Predictors
Y = df_cleaned["vadd"]  # Target variable
model = LinearRegression() # Initialize linear regression model
kf = KFold(n_splits=10, shuffle=True, random_state=42) # Perform 10-fold cross-validation
mse_scores = []
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    Y_train, Y_test = Y.iloc[train_index], Y.iloc[test_index]
    model.fit(X_train, Y_train) # Train the model
    predictions = model.predict(X_test) # Make predictions
    mse = mean_squared_error(Y_test, predictions) # Calculate Mean Squared Error
    mse_scores.append(mse)
average_mse = sum(mse_scores) / len(mse_scores) # Calculate average Mean Squared Error
print("Average Mean Squared Error:", average_mse)
```

```python
from sklearn.model_selection import KFold, learning_curve
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6)) # Plot actual vs. predicted values
plt.scatter(Y_test, predictions)
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], 'r--')
plt.xlabel('Actual VADD')
plt.ylabel('Predicted VADD')
plt.title('Actual vs. Predicted VADD')
plt.grid(True)
plt.show()
```

The scatter plot depicting actual versus predicted values of Vadd showcases a pattern where the points densely cluster around a diagonal line, indicating a close alignment between the model's predictions and the observed data. This clustering suggests that the model accurately captures the underlying relationships within the data. Because scatter plot points are close to this line, it confirms that the predictions are indeed close to the true values of Vadd.

```python
residuals = Y_test - predictions # Plot residuals
plt.figure(figsize=(10, 6))
plt.scatter(predictions, residuals)
plt.xlabel('Predicted VADD')
plt.ylabel('Residuals')
plt.title('Residuals vs. Predicted VADD')
plt.axhline(y=0, color='r', linestyle='--')
plt.grid(True)
plt.show()
```

Residuals versus Predicted Vadd, the distribution of residuals around the zero line suggests that the model's predictions are unbiased. The even spread of residuals indicates that, on average, the model does not overestimates or underestimates the true values of Vadd. Overall, this graph indicates that the model's predictions are reliable and further contributes to its effectiveness in accurately estimating the values of Vadd.

```python
train_sizes = np.arange(0.1, 1.1, 0.1) # Plot learning curve
train_sizes_abs, train_scores, test_scores = learning_curve(model, X, Y, train_sizes=train_sizes, cv=10, scoring='neg_mean_squared_error')
train_scores_mean = -np.mean(train_scores, axis=1)
test_scores_mean = -np.mean(test_scores, axis=1)
plt.figure(figsize=(10, 6))
plt.plot(train_sizes_abs, train_scores_mean, 'o-', label='Training error')
plt.plot(train_sizes_abs, test_scores_mean, 'o-', label='Validation error')
plt.xlabel('Training Size')
plt.ylabel('Mean Squared Error')
plt.title('Learning Curve')
plt.legend()
plt.grid(True)
plt.show()
```

The decreasing validation error coupled with the convergence of training and validation errors at a low value signifies that the model is effectively learning from the training data and generalizing well to unseen data. This pattern suggests that the model strikes a balance between complexity and simplicity, neither overfitting nor underfitting the data. By achieving good generalization performance, the model demonstrates its ability to accurately predict outcomes for new, unseen data points. This convergence indicates a robust and stable model, poised to perform reliably across various real-world scenarios.


## Tree-Based Model

We will use a decision tree regression to implement a tree-based model for predicting the VADD variable from the dataset. This type of model is well-suited for capturing non-linear relationships between the predictors and the target variable. For further analysis and to assess the model's performance, we will also be generating various plots of actual vs. predicted values and a residual plot as well.


```python
## Tree Based Model
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
X = df_cleaned.drop(columns=['vadd'])
Y = df_cleaned['vadd']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
tree_model = DecisionTreeRegressor(max_depth=5, random_state=42) # Initialize and fit the decision tree regressor
tree_model.fit(X_train, y_train)
tree_predictions = tree_model.predict(X_test)
```

Evaluation of model using the R^2 and Test MSE:

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, tree_predictions)
r2 = r2_score(y_test, tree_predictions)
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
